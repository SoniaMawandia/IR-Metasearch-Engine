# Task 2: Aggregated News Collection

```python
news_results1 = df1[['content','publishedAt','title','url']]
news_results1['category']= category1
news_results1['rank']= news_results1.index
news_results1['API']= 'NewsAPI'

news_results2 = df2[['content','publishedAt','title','url']]
news_results2['category']= category2
news_results2['rank']= news_results2.index
news_results2['API']= 'NewsAPI'

news_results3 = df3[['published','title','url','category']]
news_results3['category'] = category1
news_results3['content'] = ''
news_results3['rank']= news_results3.index
news_results3['API']= 'CurrentsAPI'

news_results4 = df4[['published','title','url','category']]
news_results4['category'] = category2
news_results4['content'] = ''
news_results4['rank']= news_results4.index
news_results4['API']= 'CurrentsAPI'

cols = ['publishedAt','title','url','category','content','rank','API'] 
news_results3.columns = cols
news_results4.columns = cols 

# Merging the results
news_results = news_results1.append(news_results2,ignore_index=True).append(news_results3,ignore_index=True).append(news_results4,ignore_index=True)
```

# Task 3: Ranking

### a) Approach 1: (Best Rank approach)

    Assumption: 
    1) The NewsAPI is popular news engine, hence any results provided by this news API will rank higher than the CurrentsAPI
    2) The results provided by the search engines are sorted by popularity. Hence the more popular news will have higher rank (lower number)
    

```python
#We have already provided a rank column based on order of news retreived by the news API. 
#Here we will create ranks for each news category using above assumptions

news = news_results.copy(deep=True)
```

```python
#MetaRank (x) = Min(Rankof searchengine1(x),Rankof searchengine2(x))

#step 1 - sort by ranks and API, assuming newsAPI is better
news_sorted = news.sort_values(by = ['rank','API','category'],ascending=[1,0,1])

#step 2 - remove the duplicates by removing the one coming lower in the sorted list
news_deduplicated = news_sorted[~news_sorted.duplicated(['url'],keep='first')]

#step 3 - Reindex the sorted and dedupcated results and use the index as new rank, making sure that the ran starts at 1
news_deduplicated.index = np.arange(1,len(news_deduplicated)+1)

#step 4 - assign metarank as the index of the data
news_deduplicated['metarank'] = news_deduplicated.index
news_deduplicated
```

```python
#merging the results and saving to txt file
news_deduplicated.to_csv("ResultantRanks_A1.txt", index=False)
```

### b) Approach 2:

    -->In our own approach we assume that NewsAPI crawls through all the major news provider sites and gives the best results ordered by popularity.
    -->While Currents API gets news from smaller providers and may not be very popular. Hence we rank news coming from news API as higher rank and any news coming from currentsAPI as lower rank. 
    -->In case of clashes/duplicates, we will remove the duplicate record from currents API. irrespective of its rank

```python
#step 1 - sort by ranks and API, assuming newsAPI is better
news_sorted = news.sort_values(by = ['API','rank','category'],ascending=[0,1,1])

#step 2 - remove the duplicates by removing the one coming lower in the sorted list
news_deduplicated = news_sorted[~news_sorted.duplicated(['url'],keep='first')]

#step 3 - Reindex the sorted and dedupcated results and use the index as new rank, making sure that the ran starts at 1
news_deduplicated.index = np.arange(1,len(news_deduplicated)+1)

#step 4 - assign metarank as the index of the data
news_deduplicated['metarank'] = news_deduplicated.index
```

# Retrieval Accuracy

```python
def RetreivalAcc(RankedData, GroundTruth):
    matched = pd.merge(RankedData,GroundTruth,how='inner')
    match_cnt = matched.shape[0]
    total_cnt = RankedData.shape[0]
    precision = match_cnt/total_cnt
    return precision
```

```python
ranked_A1 = pd.read_csv("ResultantRanks_A1.txt")
ranked_A2 = pd.read_csv("ResultantRanks_A2.txt")
ranked_GroundTruth = pd.read_csv("RankedDocuments.txt")
```

```python
#Precision@K
lst = []
rng = [5,10,15,20,25,30,40,50,60]
for K in rng:
    RankedData_a1 = ranked_A1[ranked_A1['metarank'] <=K]['url']
    RankedData_a2 = ranked_A2[ranked_A2['metarank'] <=K]['url']
    RankedData_GT = ranked_GroundTruth[ranked_GroundTruth['ground_truth_rank'] <=K]['url']
    a1_acc = RetreivalAcc(RankedData_a1,RankedData_GT)
    a2_acc = RetreivalAcc(RankedData_a2,RankedData_GT)
    precision = {"K":K,"A1 Accuracy":a1_acc,"A2 Accuracy":a2_acc}
    lst.append(precision)
```

```python
lst
```

```python
results_df = pd.DataFrame(lst)
results_df.head()
```
